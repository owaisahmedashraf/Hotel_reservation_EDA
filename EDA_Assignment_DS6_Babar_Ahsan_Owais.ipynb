#### Encoding

**Identify columns and then find a suitable technique for it**

```python
# prompt: give the object columns

df1.select_dtypes(include='object').columns.tolist()
```

```python
categorical_column = df1.select_dtypes(include='object').columns
categorical_column
```

```python
df1['type_of_meal_plan'].unique()
```

```python
for i in categorical_column:

  a= df1[i].unique()
  print('The categories are:',a)
  print('--------------------------------------')
  b=df1[i].value_counts()
  print(b)
  print("--------------------------------------")
```

```python
#we used one hot encoding because there is less cardinality and there was no order
df1 = pd.get_dummies(df1,columns=['type_of_meal_plan'],dtype=int)
df1.head(20)
```

```python
#we used binary encoding here because there was no order and there was moderate/high cardinality
encoder = ce.BinaryEncoder(cols=['room_type_reserved'])
df1=encoder.fit_transform(df1)
df1.head()
```

```python
#feature count encoding was used here because the value counts were unique and also only one column was added

frequency = df1['market_segment_type'].value_counts()
df1['market_segment_type_frequency_encoded']=df1['market_segment_type'].map(frequency)
df1[['market_segment_type','market_segment_type_frequency_encoded']].head()
```

```python
df1.drop(labels='market_segment_type', axis=1, inplace=True)
```
